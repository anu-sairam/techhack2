```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
dia = pd.read_csv('/content/drive/MyDrive/d1hackathon/depression.csv')
```

```python
print(dia.head())
```

```python
encoders = {col: LabelEncoder() for col in dia.select_dtypes(include=['object']).columns}
```

```python
for col, encoder in encoders.items():
    dia[col] = encoder.fit_transform(dia[col])
```

```python
X = dia.drop('Depression', axis=1)
y = dia['Depression']
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
X = imputer.fit_transform(X)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X, y)
```

```python
def encode_input(data, encoders):
    encoded_data = []
    for col, value in data.items():
        if col in encoders:
            encoded_data.append(encoders[col].transform([value])[0])
        else:
            encoded_data.append(value)
    return np.array(encoded_data).reshape(1, -1)
```

```python
input_data = {
    "id": int(input("Enter the id: ")),
    "Age": int(input("Enter the age: ")),
    "City": input("Enter the city: "),
    "Profession": input("Enter the profession: "),
    "Academic Pressure": int(input("Enter the academic pressure: ")),
    "Work Pressure": int(input("Enter the work pressure: ")),
    "CGPA": float(input("Enter the CGPA: ")),
    "Study Satisfaction": int(input("Enter the study satisfaction: ")),
    "Job Satisfaction": int(input("Enter the job satisfaction: ")),
    "Sleep Duration": input("Enter the sleep duration: "),
    "Dietary Habits": input("Enter the dietary habits: "),
    "Degree": input("Enter the degree: "),
    "Have you ever had suicidal thoughts ?": input("Have you ever had suicidal thoughts (Yes/No)?: "),
    "Work/Study Hours": int(input("Enter the work/study hours: ")),
    "Financial Stress": float(input("Enter the financial stress: ")),
    "Family History of Mental Illness": input("Enter the family history of mental illness (Yes/No): "),
}
```

```python
try:
    encoded_data = encode_input(input_data, encoders)
    # Predict using the trained model
    prediction = lr.predict(encoded_data)
    print(f"Prediction (Depression: 1=Yes, 0=No): {prediction[0]}")
except ValueError as e:
    print(f"Result: {e}")
```
